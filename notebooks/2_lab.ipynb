```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
ds = load_iris()
ds
```

```python
df = pd.DataFrame(ds.data, ds.target, columns=ds.feature_names)
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(ds.data, ds.target, train_size=0.8, random_state=42)
print(f'x_train shape: {x_train.shape}, x_test shape: {x_test.shape}')
print(f'y_test shape: {y_train.shape}, y_test shape: {y_test.shape}')
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(x_train, y_train)
print("accuracy on train set {:.2f}".format(clf.score(x_train, y_train)))
print("accuracy on test set {:.2f}".format(clf.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = clf.predict(x_train)
cm = confusion_matrix(y_train, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```

```python
from imblearn.metrics import specificity_score, sensitivity_score


def show_classification_characteristics(clf, x, y):
    print("accuracy: {:.2f}".format(clf.score(x, y)))
    print("error rate: {:.2f}".format(1 - clf.score(x, y)))
    y_pred = clf.predict(x)
    print("sensitivity")
    print(sensitivity_score(y, y_pred, average=None))
    print("specificity")
    print(specificity_score(y, y_pred, average=None))


show_classification_characteristics(clf, x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```

```python
show_classification_characteristics(clf, x_test, y_test)
```

# Подбор гипер параметров

```python
ds = load_iris()
df = pd.DataFrame(ds.data, ds.target, columns=ds.feature_names)
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(ds.data, ds.target, train_size=0.8, random_state=42)
```

```python
naive_bayes_train_scores = []
naive_bayes_test_scores = []
interval = range(1, 1001)
for alpla in interval:
    clf = MultinomialNB(alpha=alpla / 100.)
    clf.fit(x_train, y_train)
    naive_bayes_train_scores.append(clf.score(x_train, y_train))
    naive_bayes_test_scores.append(clf.score(x_test, y_test))
```

```python
# xs = [x / 100. for x in interval]
xs = np.linspace(0, 10, 1000)
plt.plot(xs, naive_bayes_train_scores, label='train scores')
plt.plot(xs, naive_bayes_test_scores, label='test scores')

plt.title('Naive Bias hyper parameters')
plt.ylim(0.6, 1.1)
plt.legend()
```

```python
decision_tree_criterias = ['gini', 'entropy', 'log_loss']
decision_tree_idxs = np.linspace(0, 2, 3, dtype=int)
decision_tree_train_scores = []
decision_tree_test_scores = []

for idx in decision_tree_idxs:
    clf = DecisionTreeClassifier(criterion=decision_tree_criterias[idx])
    clf.fit(x_train, y_train)
    decision_tree_train_scores.append(clf.score(x_train, y_train))
    decision_tree_test_scores.append(clf.score(x_test, y_test))
```

```python
plt.plot(decision_tree_idxs, decision_tree_train_scores, label='train scores', linewidth=5)
plt.plot(decision_tree_idxs, decision_tree_test_scores, label='test scores')

plt.title('Decision tree hyper parameters')
plt.ylim(0.95, 1.01)
plt.legend()
```

```python
svc_c_parameters = np.linspace(0.1, 10, 100)
svc_train_scores = []
svc_test_scores = []

for i in range(100):
    clf = SVC(C=svc_c_parameters[i])
    clf.fit(x_train, y_train)
    svc_train_scores.append(clf.score(x_train, y_train))
    svc_test_scores.append(clf.score(x_test, y_test))
```

```python
plt.plot(svc_c_parameters, svc_train_scores, label='train scores')
plt.plot(svc_c_parameters, svc_test_scores, label='test scores')

plt.title('Decision tree hyper parameters')
plt.ylim(0.7, 1.1)
plt.legend()
```

# Подбор гипер-параметров 2 (По-нормальному)

```python
ds = load_iris()
df = pd.DataFrame(ds.data, ds.target, columns=ds.feature_names)
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(ds.data, ds.target, train_size=0.8, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha': [0.1, 0.5, 1.0],
    'fit_prior': [True, False]
}

nb = MultinomialNB()

cv = GridSearchCV(estimator=nb, param_grid=param_grid, cv=5)
cv.fit(x_train, y_train)
print('best estimator params {}'.format(cv.best_estimator_))
print('best estimator score {:.2f}'.format(cv.best_score_))
```

```python
from pprint import pprint

results = cv.cv_results_
pprint(cv.cv_results_)
param_alpha = results['param_alpha']
param_fit_prior = results['param_fit_prior']
mean_test_score = results['mean_test_score']

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(param_alpha, param_fit_prior, mean_test_score, c=mean_test_score, cmap='viridis')
ax.set_xlabel('Alpha')
ax.set_ylabel('Fit Prior')
ax.set_zlabel('Mean Test Score')
ax.set_title('Grid Search Results')

sc = ax.scatter(param_alpha, param_fit_prior, mean_test_score, c=mean_test_score, cmap='viridis')
fig.colorbar(sc, ax=ax, label='Mean Test Score')

plt.show()
```

```python
import seaborn as sns

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': np.arange(1, 21)
}

dt = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

scores = grid_search.cv_results_['mean_test_score']
scores = scores.reshape(len(param_grid['criterion']),
                        len(param_grid['max_depth']),
                        len(param_grid['min_samples_split']),
                        len(param_grid['min_samples_leaf']))

criterion_values = param_grid['criterion']
max_depth_values = param_grid['max_depth']
min_samples_split_values = param_grid['min_samples_split']
min_samples_leaf_values = param_grid['min_samples_leaf']

plt.figure(figsize=(14, 10))
sns.heatmap(scores.mean(axis=(0, 3)), annot=True, fmt=".3f", cmap='YlGnBu',
            xticklabels=min_samples_split_values, yticklabels=max_depth_values,
            cbar_kws={'label': 'Mean Test Score'})
plt.xlabel('min_samples_split')
plt.ylabel('max_depth')
plt.title(f'Grid Search Results (Criterion: {criterion_values[0]})')
plt.show()
```

```python
param_grid = {
    'C': [0.1, 1.0, 10.0],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.1, 1.0]
}

svc = SVC()

cv = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5)
cv.fit(x_train, y_train)
print('best estimator params {}'.format(cv.best_estimator_))
print('best estimator score {:.2f}'.format(cv.best_score_))

results = cv.cv_results_
param_C = results['param_C']
param_kernel = results['param_kernel']
param_gamma = results['param_gamma']
mean_test_score = results['mean_test_score']

param_C_map = {0.1: 0, 1.0: 1, 10.0: 2}
param_kernel_map = {'linear': 0, 'rbf': 1, 'poly': 2}
param_gamma_map = {'scale': 0, 'auto': 1, 0.1: 2, 1.0: 3}

param_C_numeric = np.array([param_C_map[x] for x in param_C])
param_kernel_numeric = np.array([param_kernel_map[x] for x in param_kernel])
param_gamma_numeric = np.array([param_gamma_map[x] for x in param_gamma])

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(param_C_numeric, param_gamma_numeric, param_kernel_numeric, c=mean_test_score, cmap='viridis')
ax.set_xlabel('C')
ax.set_ylabel('Gamma')
ax.set_zlabel('Kernel')
ax.set_title('Grid Search Results')

sc = ax.scatter(param_C_numeric, param_gamma_numeric, param_kernel_numeric, c=mean_test_score, cmap='viridis')
fig.colorbar(sc, ax=ax, label='Mean Test Score')

plt.show()
```
